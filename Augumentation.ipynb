```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import random
from PIL import Image, ImageEnhance

def cv2_clipped_zoom(img, zoom_factor):
    height, width = img.shape[:2] # It's also the final desired shape
    new_height, new_width = int(height * zoom_factor), int(width * zoom_factor)
    y1, x1 = max(0, new_height - height) // 2, max(0, new_width - width) // 2
    y2, x2 = y1 + height, x1 + width
    bbox = np.array([y1,x1,y2,x2])
    # Map back to original image coordinates
    bbox = (bbox / zoom_factor).astype(np.int)
    y1, x1, y2, x2 = bbox
    cropped_img = img[y1:y2, x1:x2]

    # Handle padding when downscaling
    resize_height, resize_width = min(new_height, height), min(new_width, width)
    pad_height1, pad_width1 = (height - resize_height) // 2, (width - resize_width) //2
    pad_height2, pad_width2 = (height - resize_height) - pad_height1, (width - resize_width) - pad_width1
    pad_spec = [(pad_height1, pad_height2), (pad_width1, pad_width2)] + [(0,0)] * (img.ndim - 2)

    result = cv.resize(cropped_img, (resize_width, resize_height))
    result = np.pad(result, pad_spec, mode='constant')
    assert result.shape[0] == height and result.shape[1] == width
    return result

def aumgent_8(impath):
  img = cv.imread(impath,1)
  im1 = cv.flip(img, 0) #Vertical Flip
  im2 = cv.flip(img, 1) #Horizontal Flip
  im3 = cv.flip(img, -1) #Flip Both
  im4 = cv2_clipped_zoom(img, random.uniform(0.9,1.2))
  im5 = cv2_clipped_zoom(im1, random.uniform(0.9,1.2))
  file_name = impath.split('/')
  new_path = file_name[-1].split('-') 
  cv.imwrite("Augument Images/"+new_path[0]+"VF-"+new_path[1],im1)
  cv.imwrite("Augument Images/"+new_path[0]+"HF-"+new_path[1],im2)
  cv.imwrite("Augument Images/"+new_path[0]+"BF-"+new_path[1],im3)
  cv.imwrite("Augument Images/"+new_path[0]+"ZM-"+new_path[1],im4)
  cv.imwrite("Augument Images/"+new_path[0]+"VFZM-"+new_path[1],im5)
  #image random rotating and brightening using PIL
  img_org = Image.open(impath)
  enhancer = ImageEnhance.Brightness(img_org)
  im6 = enhancer.enhance(random.uniform(1.0,1.3))
  im7 = img_org.rotate(random.uniform(0,10))
  img_ver = Image.open("Augument Images/"+new_path[0]+"VF-"+new_path[1])
  im8 = img_ver.rotate(random.uniform(0,10))
  im6.save("Augument Images/"+new_path[0]+"RB-"+new_path[1])
  im7.save("Augument Images/"+new_path[0]+"RR-"+new_path[1])
  im8.save("Augument Images/"+new_path[0]+"VFRR-"+new_path[1])
  print("saved 8")
    
def aumgent_11(impath):
  img = cv.imread(impath,1)
  im1 = cv.flip(img, 0) #Vertical Flip
  im2 = cv.flip(img, 1) #Horizontal Flip
  im3 = cv.flip(img, -1) #Both Flip
  im4 = cv2_clipped_zoom(img, random.uniform(0.9,1.2))
  im5 = cv2_clipped_zoom(im1, random.uniform(0.9,1.2))
  im6 = cv2_clipped_zoom(im2, random.uniform(0.9,1.2))
  im7 = cv2_clipped_zoom(im3, random.uniform(0.9,1.2))
  file_name = impath.split('/')
  new_path = file_name[-1].split('-') 
  cv.imwrite("Augument Images/"+new_path[0]+"VF-"+new_path[1],im1)
  cv.imwrite("Augument Images/"+new_path[0]+"HF-"+new_path[1],im2)
  cv.imwrite("Augument Images/"+new_path[0]+"BF-"+new_path[1],im3)
  cv.imwrite("Augument Images/"+new_path[0]+"ZM-"+new_path[1],im4)
  cv.imwrite("Augument Images/"+new_path[0]+"VFZM-"+new_path[1],im5)
  cv.imwrite("Augument Images/"+new_path[0]+"HFZM-"+new_path[1],im6)
  cv.imwrite("Augument Images/"+new_path[0]+"BFZM-"+new_path[1],im7)
  #image random rotating and brightening using PIL
  img_org = Image.open(impath)
  enhancer = ImageEnhance.Brightness(img_org)
  im7 = enhancer.enhance(random.uniform(0.9,1.2))
  im8 = img_org.rotate(random.uniform(0,10))
  img_ver = Image.open("Augument Images/"+new_path[0]+"VF-"+new_path[1])
  im9 = img_ver.rotate(random.uniform(0,10))
  im10 = ImageEnhance.Brightness(im9).enhance(random.uniform(0.9,1.2))
  im11 =  ImageEnhance.Brightness(img_ver).enhance(random.uniform(0.9,1.2))
  im7.save("Augument Images/"+new_path[0]+"RB-"+new_path[1])
  im8.save("Augument Images/"+new_path[0]+"RR-"+new_path[1])
  im9.save("Augument Images/"+new_path[0]+"VFRR-"+new_path[1])
  im10.save("Augument Images/"+new_path[0]+"VFRRRB-"+new_path[1])
  im11.save("Augument Images/"+new_path[0]+"VFRB-"+new_path[1])
  print("saved 11")
    
#for H class that has the smallest set
def aumgent_20(impath):
  img = cv.imread(impath,1)
  im1 = cv.flip(img, 0) #Vertical Flip
  im2 = cv.flip(img, 1) #Horizontal Flip
  im3 = cv.flip(img, -1) #Both Flip
  im4 = cv2_clipped_zoom(img, random.uniform(0.9,1.2))
  im5 = cv2_clipped_zoom(im1, random.uniform(0.9,1.2))
  im6 = cv2_clipped_zoom(im2, random.uniform(0.9,1.2))
  im7 = cv2_clipped_zoom(im3, random.uniform(0.9,1.2))
  file_name = impath.split('/')
  new_path = file_name[-1].split('-') 
  cv.imwrite("Augument Images/"+new_path[0]+"VF-"+new_path[1],im1)
  cv.imwrite("Augument Images/"+new_path[0]+"HF-"+new_path[1],im2)
  cv.imwrite("Augument Images/"+new_path[0]+"BF-"+new_path[1],im3)
  cv.imwrite("Augument Images/"+new_path[0]+"ZM-"+new_path[1],im4)
  cv.imwrite("Augument Images/"+new_path[0]+"VFZM-"+new_path[1],im5)
  cv.imwrite("Augument Images/"+new_path[0]+"HFZM-"+new_path[1],im6)
  cv.imwrite("Augument Images/"+new_path[0]+"BFZM-"+new_path[1],im7)
  #image random rotating and brightening using PIL
  img_org = Image.open(impath)
  enhancer = ImageEnhance.Brightness(img_org)
  im7 = enhancer.enhance(random.uniform(0.9,1.2))
  im8 = img_org.rotate(random.uniform(0,10))
  img_ver = Image.open("Augument Images/"+new_path[0]+"VF-"+new_path[1])
  im9 = img_ver.rotate(random.uniform(0,10))
  im10 = ImageEnhance.Brightness(im9).enhance(random.uniform(0.9,1.2))
  im11 =  ImageEnhance.Brightness(img_ver).enhance(random.uniform(0.9,1.2))
  img_hor = Image.open("Augument Images/"+new_path[0]+"HF-"+new_path[1])
  im12 = img_hor.rotate(random.uniform(0,10))
  im13 = ImageEnhance.Brightness(im12).enhance(random.uniform(0.9,1.2))
  im14 = ImageEnhance.Brightness(img_hor).enhance(random.uniform(0.9,1.2))
  im_bfl =  Image.open("Augument Images/"+new_path[0]+"BF-"+new_path[1])
  im15 = im_bfl.rotate(random.uniform(0,10))
  im16 = ImageEnhance.Brightness(im15).enhance(random.uniform(0.9,1.2))
  im17 = ImageEnhance.Brightness(im_bfl).enhance(random.uniform(0.9,1.2))
  im_zom =  Image.open("Augument Images/"+new_path[0]+"ZM-"+new_path[1])
  im18 = im_zom.rotate(random.uniform(0,10))
  im19 = ImageEnhance.Brightness(im18).enhance(random.uniform(0.9,1.2))
  im20 = ImageEnhance.Brightness(im_zom).enhance(random.uniform(0.9,1.2))

  im7.save("Augument Images/"+new_path[0]+"RB-"+new_path[1])
  im8.save("Augument Images/"+new_path[0]+"RR-"+new_path[1])
  im9.save("Augument Images/"+new_path[0]+"VFRR-"+new_path[1])
  im10.save("Augument Images/"+new_path[0]+"VFRRRB-"+new_path[1])
  im11.save("Augument Images/"+new_path[0]+"VFRB-"+new_path[1])
  im12.save("Augument Images/"+new_path[0]+"HFRR-"+new_path[1])
  im13.save("Augument Images/"+new_path[0]+"HFRRRB-"+new_path[1])
  im14.save("Augument Images/"+new_path[0]+"HFRB-"+new_path[1])
  im15.save("Augument Images/"+new_path[0]+"BFRR-"+new_path[1])
  im16.save("Augument Images/"+new_path[0]+"BFRRRB-"+new_path[1])
  im17.save("Augument Images/"+new_path[0]+"BFRB-"+new_path[1])
  im18.save("Augument Images/"+new_path[0]+"ZMRR-"+new_path[1])
  im19.save("Augument Images/"+new_path[0]+"ZMRRRB-"+new_path[1])
  im20.save("Augument Images/"+new_path[0]+"ZMRB-"+new_path[1])
  print("saved 20")
```
